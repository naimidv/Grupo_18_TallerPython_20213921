1. Copia el siguiente diccionario en una c√©dula de Colab. A partir de √©l, crea un dataframe
llamado ùíèùíá o utilizando Pandas.

```python
data_ciudades = {'Ciudad': ['Lima', 'Arequipa', 'Trujillo', 'Cusco', 'Chiclayo', 'Piura', 'Iquitos', 'Huancayo', 'Tacna', 'Pucallpa'], 'Habitantes':[1047996, 100169, 92331, 428450, 305717, 484475, 441649, 385098, 294395, 283734], 'Area_km2':[2672.28, 1545.77, 1487.7, 116.5, 279.89, 6217.26, 368.9, 109.19, 59.4, 483.44], 'Altitud_m':[154, 2325, 34, 3399, 29, 29, 106, 3271, 562, 156], 'Densidad_poblacion_por_km2':[3924, 64.8, 62.1, 3673, 1091.6, 77.9, 1196, 3528.6, 4966, 587.2]}

nf = pd.DataFrame(data_ciudades)
nf
```

2. Realiza el mismo procedimiento con el siguiente diccionario. A partir de √©l, crea un
dataframe llamado ùíèùíêùíéùíÉùíìùíÜùíî.

```python
data_ciudades_2= {'Ciudad':['Lima', 'Arequipa', 'Trujillo', 'Cusco', 'Chiclayo', 'Piura', 'Iquitos', 'Huancayo', 'Tacna', 'Pucallpa'], 'Gentilicio': ['Limense', 'Arequipe√±o', 'Trujillano', 'Cusque√±o', 'Chiclayano', 'Piurano', 'Iquite√±o', 'Huancaino', 'Tacne√±o', 'Pucallpino'], 'Provincia': ['Lima', 'Arequipa', 'Trujillo', 'Cusco', 'Chiclayo', 'Piura', 'Maynas', 'Huancayo', 'Tacna', 'Coronel Portillo'], 'Region':['Lima', 'Arequipa', 'La Libertad', 'Cusco', 'Lambayeque', 'Piura', 'Loreto', 'Junin', 'Tacna', 'Ucayali'] }
nombres = pd.DataFrame(data_ciudades_2)
nombres
```

3. Con ambos dataframes, realiza un inner join. Guarda el resultado en un nuevo dataframe
llamado ùíÑùíñùíÇùíÖùíìùíê_ùüè


```python
cuadro_1 = pd.merge(nf, nombres, on='Ciudad', how='inner')
cuadro_1
```

4. Resume las estad√≠sticas descriptivas del dataframe y responde ¬øCu√°l es la m√≠nima
densidad poblacional? ¬øA cu√°l ciudad corresponde? ¬øY la m√°xima?

```python
cuadro_1.describe()
```

```python
densidad_minima = cuadro_1['Densidad_poblacion_por_km2'].min()
ciudad_minima = cuadro_1.loc[cuadro_1['Densidad_poblacion_por_km2']==densidad_minima, 'Ciudad'].values[0]

densidad_maxima = cuadro_1['Densidad_poblacion_por_km2'].max()
ciudad_maxima = cuadro_1.loc[cuadro_1['Densidad_poblacion_por_km2']==densidad_maxima, 'Ciudad'].values[0]

print(f"La m√≠nima densidad poblacional es {densidad_minima} y corresponde a la ciudad de {ciudad_minima}.")
print(f"La m√°xima densidad poblacional es {densidad_maxima} y corresponde a la ciudad de {ciudad_maxima}.")
```

5. Realizar un gr√°fico de barras donde se vea la cantidad de habitantes en cada ciudad

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.bar(x = cuadro_1["Ciudad"], height= cuadro_1["Habitantes"], color="steelblue")
ax.set_xlabel('Ciudad')
ax.set_ylabel('Habitantes')
ax.set_title('N√∫mero de Habitantes por Ciudad en Per√∫')
ax.tick_params(axis="x", rotation=45)
plt.show()
```

6. Realizar un gr√°fico de dispersi√≥n entre la altura y el n√∫mero de habitantes en las
ciudades del dataframe

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(cuadro_1['Altitud_m'], cuadro_1['Habitantes'], color='green')
ax.set_title('Relaci√≥n entre Altitud y N√∫mero de Habitantes de las Ciudades en Per√∫')
ax.set_xlabel('Altitud (m)')
ax.set_ylabel('Habitantes')

ax.grid(True)

plt.show()
```
