5. Utilizando ùíîùíÑùíäùíëùíö. ùíîùíïùíÇùíïùíî, realiza un an√°lisis de una distribuci√≥n normal con una media de
0 y desviaci√≥n est√°ndar de 1. Genera un conjunto de datos aleatorios con 1000
muestras a partir de esta distribuci√≥n.

```python
import numpy as np
from scipy.stats import norm
```

```python
mean = 0
std = 1

datos = norm.rvs(mean, std, size=1000)
```

6. Calcula y muestra la media, la desviaci√≥n est√°ndar y la mediana de los datos
generados.


```python
mean = np.mean(datos)
print(f'La media de los datos generados es {mean}')

std = np.std(datos)
print(f'La desviaci√≥n est√°ndar de los datos generados es {std}')

median = np.median(datos)
print(f'La mediana de los datos generados es {median}')
```

7. Grafica el histograma de los datos y superp√≥n la curva de la distribuci√≥n normal
te√≥rica.


```python
figs, ax = plt.subplots(figsize=(6.4,6))

ax.hist(datos, bins=40, density= True, color='green')

xmin, xmax = ax.get_xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std)
ax.plot(x, p, 'k', linewidth=2)


plt.title('Histograma de datos generados y curva de distribuci√≥n normal')
plt.xlabel('Valor')
plt.ylabel('Densidad de probabilidad')

plt.show()
```

7. Calcula la probabilidad de que un valor caiga dentro del rango de -1 a 1 (una desviaci√≥n
est√°ndar de la media).

```python
prob = norm.cdf(1, mean, std) - norm.cdf(-1, mean, std)
print(f"La probabilidad de que un valor caiga dentro del rango de -1 y 1 es {prob}")
```
