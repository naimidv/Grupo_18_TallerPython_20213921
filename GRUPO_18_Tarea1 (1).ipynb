III. Crea una funci√≥n ùíÑùíÇùíçùíÑùíñùíçùíÇùíì_ùíÖùíÜùíîùíÑùíñùíÜùíèùíïùíê() que permita calcular el descuento aplicado a un
producto.

```python
def calcular_descuento(tot):
    if tot>=1000:
        desc=0.2
    elif tot>=500 and tot<1000:
        desc=0.1
    else:
        desc=0
    return tot-(tot*desc)
```

```python
totales=[1200, 750, 300]
for tot in totales:
    print(calcular_descuento(tot))
```

```python
def clasificar_edad(edad):
    if edad < 12:
        return "Ni√±o"
    elif 12 <= edad <= 17:
        return "Adolescente"
    elif 18 <= edad <= 24:
        return "Joven"
    elif 25 <= edad <= 64:
        return "Adulto"
    elif edad >= 65:
        return "Adulto Mayor"
```

```python
edades = [10, 19, 25, 70]
for edad in edades:
    print(clasificar_edad(edad))
```

```python
def calcular_craest(media_personal, media_curso, desviacion_curso):
    craest = ((media_personal - media_curso) * 10) / desviacion_curso + 50
    return craest
```
